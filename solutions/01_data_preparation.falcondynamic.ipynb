```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
from collections import Counter
```

```python
# open and read txt file 
with open('../data/human_chat.txt', 'r', encoding='utf-8') as file:
    data = file.read()

# Preview the content by printing the first 500 characters of the file
print(data[:500])
```

```python
# Divide lines
lines = data.split('\n')   
print(lines[:10])  
```

```python
#removing Human 1: or Human 2: from the begining of the line
lines = [re.sub(r"^Human \d+: ", "", line) for line in lines if line.strip() != ""]
print(lines[:10])
```

```python
#Creating WordCloud
# merging lines
text = ' '.join(lines)

# creating word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

#Word Cloud Visualisation
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Most Common Words
words = re.findall(r'\w+', text.lower())
print('words:', words)
word_freq = Counter(words)
top_words = word_freq.most_common(30)
print('top_words:', top_words)
```
